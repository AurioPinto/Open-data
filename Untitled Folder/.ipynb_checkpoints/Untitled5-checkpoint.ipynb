```python
import math

EARTH_REDIUS = 6378.137
pi=3.14
level1=10
level2=15
level3=25
level4=35



def rad(d):
    return d * pi / 180.0



def getDistance(lat1, lng1, lat2, lng2):
    radLat1 = rad(lat1)
    radLat2 = rad(lat2)
    a = math.rad(lat1 - lat2)
    b = math.rad(lng1 - lng2)
    s = 2 * math.asin(math.sqrt(math.pow(math.sin(a/2), 2) + math.cos(radLat1) * math.cos(radLat2) * math.pow(math.sin(b/2), 2)))
    s = s * EARTH_REDIUS
    return s





import pandas as pd
import math
import matplotlib.pyplot as plt
import time
import numpy as np
from sklearn.cluster import DBSCAN

file=pd.read_csv('C:\\Users\\win\Documents\\Thesis\\Data\\source data\\gps_20161101.csv')
df=pd.DataFrame(file)
df['velosity']=None
df['CrowdLevel']=None
df['tm_wday']=None#
df['tm_hour']=None

df.iloc[[]]
lng1=df.iloc[0,3]
lat1=df.iloc[0,4]
time1=df.iloc[0,2]
lng2=df.iloc[0,3]
lat2=df.iloc[0,4]
time2=0
print(df.iloc[0])


for row in df.itertuples():
    print(getattr(row,'Index'))
    lng1=lng2
    lat1=lat2
    time1=time2
    lng2=getattr(row, 'longitude')
    lat2=getattr(row, 'latitude')
    time2=getattr(row,'time')
   
    velosity= abs(getDistance(lat1,lng1,lat2,lng2)/(time2-time1))*3600
    df.iloc[getattr(row,'Index'),5]=velosity
    if velosity>level4:
        df.iloc[getattr(row,'Index'),6]=4
    elif velosity>level3:
        df.iloc[getattr(row,'Index'),6]=3
    elif velosity>level2:
        df.iloc[getattr(row,'Index'),6]=2
    elif velosity>level1:
        df.iloc[getattr(row,'Index'),6]=1
    else:
        df.iloc[getattr(row,'Index'),6]=0
    timestamp = getattr(row, 'time')
    time_local = time.localtime(timestamp)
    df.iloc[getattr(row,'Index'),7]=time_local.tm_wday
    df.iloc[getattr(row,'Index'),8]=time_local.tm_hour 

df.to_csv('C:\\Users\\win\Documents\\Thesis\\Data\\source data\\gps_20161101.csv',encoding='gbk')
 


    
import pandas as pd
import math
import matplotlib.pyplot as plt
import time
import numpy as np
from sklearn.cluster import DBSCAN



file=pd.read_csv('‪C:\\Users\\win\\Documents\\Thesis\\Data\\all.csv',low_memory=False)
df1=pd.DataFrame(file)
df2=df1[df1['CrowdLevel']==0]
df2=df2.reset_index()

cls = DBSCAN(eps=0.0001,min_samples=1).fit(df2[['longitude','latitude']].values)

n_clusters = len(set(cls.labels_))

labels = pd.DataFrame(cls.labels_)
res = pd.concat([df2,labels],axis=1,ignore_index=True)
mylist=res[11].tolist()
res['Zone']=None
myset=set(mylist)
Num=[]
for item in myset:
    Num.append(mylist.count(item))
myset=list(myset)    
c ={'myset':myset,'Num':Num}
df_Ana =pd.DataFrame(c)
for item in df_Ana[df_Ana['Num']>500]['myset']:
    res.loc[res[11]==item,'Zone']=1

res.columns = ['index1', 'index2','driver','order','time','longitude','latitude','speed','CrowdLevel','wday','hour','labels','CrowdZone']

res.to_csv('‪C:\\Users\\win\\Documents\\Thesis\\Data\\result.csv',encoding='gbk')



file=pd.read_csv('‪C:\\Users\\win\\Documents\\Thesis\\Data\\result.csv',low_memory=False)
res=pd.DataFrame(file)
Croad=res[res['CrowdZone']==1].drop_duplicates(subset='labels')['labels'].tolist()



for item in Croad:
    plt.scatter(res.loc[res['labels']==item]['longitude'],res.loc[res['labels']==item]['latitude'],s=60,c='b',alpha=0.5) 
    plt.show()

    
    
    
    
    
    
    import pandas as pd
import pandas as pd
import math
import matplotlib.pyplot as plt
import time
import numpy as np
from sklearn.cluster import DBSCAN


file=pd.read_csv( '‪C:\\Users\\win\\Documents\\Thesis\\Data\\result.csv',low_memory=False)
df1=pd.DataFrame(file)
df2=df1[df1['CrowdZone']==1]
df2=df2.reset_index()
df2.columns = ['index1', 'index2', 'index3', 'index4','driver','order','time','longitude','latitude','speed','CrowdLevel','wday','hour','labels','CrowdZone']
df2.to_csv('‪C:\\Users\\win\\Documents\\Thesis\\Data\\model.csv',encoding='gbk')




file=pd.read_csv('‪C:\\Users\\win\\Documents\\Thesis\\Data\\model.csv',low_memory=False)
model=pd.DataFrame(file)

model['Pre_Time']=None


model.drop_duplicates(subset={'wday','hour','labels'})
for item1 in model.drop_duplicates(subset='labels')['labels']:
    Temp=model[model['labels']==item1]
    for item2 in Temp.drop_duplicates(subset='wday')['wday']:
        DayData=Temp[Temp['wday']==item2].drop_duplicates(subset='hour')['hour'].tolist()
        DayData.sort()
        for i in range(len(DayData)):
            for j in range(len(DayData)-i):
                if DayData[i]+j!=DayData[i+j]:
                    model.loc[(model['labels']==item1) & (model['wday']==item2) & (model['hour']==DayData[i]),'Pre_Time']=j
                    break
    
model.to_csv('‪C:\\Users\\win\\Documents\\Thesis\\Data\\model.csv',encoding='gbk')





file=pd.read_csv('‪C:\\Users\\win\\Documents\\Thesis\\Data\\model.csv',low_memory=False)
model=pd.DataFrame(file)
lat=30.67222
lng=104.05914
hour=15
wday=1
for row in model.itertuples():
    print(getattr(row,'Index'))
    if getDistance(lat, lng, getattr(row,'latitude'), getattr(row,'longitude'))<100:
        RoadNum=getattr(row,'labels')
        Pre_Time=getattr(row,'Pre_Time')
        print('You are in the road of '+str(RoadNum)+' and the predict time is '+str(Pre_Time))
        break
     
    
```
